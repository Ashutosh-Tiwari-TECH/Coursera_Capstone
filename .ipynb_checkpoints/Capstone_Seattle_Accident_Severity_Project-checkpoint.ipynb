```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
mdl_data=pd.DataFrame(mdl_data)
mdl_data = mdl_data.dropna()
mdl_data = mdl_data.reset_index(drop=True)
X=mdl_data[['ADDRTYPE','COLLISIONTYPE','WEATHER','ROADCOND','LIGHTCOND']].values

le_sex = preprocessing.LabelEncoder()
le_sex.fit(['Block','Intersection'])
X[:,0] = le_sex.transform(X[:,0])

le_Col = preprocessing.LabelEncoder()
le_Col.fit(['Angles','Sideswipe','Parked Car','Other','Cycles','Rear Ended','Head On','Left Turn','Pedestrian','Right Turn'])
X[:,1] = le_Col.transform(X[:,1])

le_Wthr = preprocessing.LabelEncoder()
le_Wthr.fit(['Overcast','Raining','Clear','Unknown','Other','Snowing','Fog/Smog/Smoke','Sleet/Hail/Freezing Rain','Blowing Sand/Dirt','Severe Crosswind','Partly Cloudy'])
X[:,2] = le_Wthr.transform(X[:,2])

le_rod = preprocessing.LabelEncoder()
le_rod.fit(['Wet','Dry','Unknown','Snow/Slush','Ice','Other','Sand/Mud/Dirt','Standing Water','Oil'])
X[:,3] = le_rod.transform(X[:,3])

le_lgt = preprocessing.LabelEncoder()
le_lgt.fit(['Daylight','Dark - Street Lights On','Dark - No Street Lights','Unknown','Dusk','Dawn','Dark - Street Lights Off','Other','Dark - Unknown Lighting'])
X[:,4] = le_lgt.transform(X[:,4])

y=mdl_data['SEVERITYCODE']

X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
drugTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
drugTree # it shows the default parameters
```

```python
drugTree.fit(X_trainset,y_trainset)
```

```python
predTree = drugTree.predict(X_testset)
```

```python
print (predTree [0:5])
print (y_testset [0:5])
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset, predTree))
```

```python
Data=pd.DataFrame(X)
Data.columns=['ADDRTYPE','COLLISIONTYPE','WEATHER','ROADCOND','LIGHTCOND']
Data['SEVERITY']=y
convrt1 = {'SEVERITY':int}
convrt = {'ADDRTYPE':int,'COLLISIONTYPE':int,'WEATHER':int,'ROADCOND':int,'LIGHTCOND':int}
Data=Data.astype(convrt1)
Data=Data.astype(convrt)

viz = Data[['ADDRTYPE','COLLISIONTYPE','WEATHER','ROADCOND','LIGHTCOND','SEVERITY']]
viz.hist()
plt.show()
```

```python
msk = np.random.rand(len(Data)) < 0.8
train = Data[msk]
test = Data[~msk]
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['COLLISIONTYPE']])
train_y = np.asanyarray(train[['SEVERITY']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['COLLISIONTYPE']])
test_y = np.asanyarray(test[['SEVERITY']])
test_y_hat = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_hat - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_hat - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_hat , test_y) )
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, mdl_data['SEVERITYCODE'], test_size=0.3, random_state=3)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k = 4
#Train Model and Predict  
neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)
neigh
```

```python
yhat = neigh.predict(X_test)
yhat[0:5]
```

```python
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(y_train, neigh.predict(X_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat))
```

```python
X = np.asarray(Data[['ADDRTYPE','COLLISIONTYPE','WEATHER','ROADCOND','LIGHTCOND']])
X[0:5]
```

```python
y = np.asarray(Data[['SEVERITY']])
y [0:5]
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
LR
```

```python
yhat = LR.predict(X_test)
yhat
```

```python
yhat_prob = LR.predict_proba(X_test)
yhat_prob
```

```python
from sklearn.metrics import jaccard_similarity_score
jaccard_similarity_score(y_test, yhat)
```
